9. Comparar três números

```python
numero1 = int( input(" Digite o numero 1: ") )
numero2 = int( input(" Digite o numero 2: ") )
numero3 = int( input(" Digite o numero 3: ") )
if numero1 > numero2 and numero1 > numero3:
    print("Numero 1 é maior que numero 2 e numero 3")
elif numero2 > numero3 and numero2 > numero1:
    print("Numero 2 é maior que numero 1 e numero 3")
else:
    print("Numero 3 é maior que numero 1 e numero 2")
```
